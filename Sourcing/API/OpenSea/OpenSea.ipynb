```python
import requests
import pandas as pd
```

```python
url = "https://api.opensea.io/api/v1/events"

querystring = {"only_opensea":"false","offset":"0","limit":"300"}

headers = {"Accept": "application/json"}

response = requests.request("GET", url, headers=headers, params=querystring)

output = response.json()

transactions = pd.DataFrame(data=output["asset_events"])
```

```python
for column_name in transactions.columns:
    print(column_name)
```

```python
transactions = transactions.drop(transactions[transactions["ending_price"].isnull() == 1].index)
```

```python
transactions["ending_price"] = transactions["ending_price"].astype(int)
```
